## Exercici 1
Descarrega el dataset adjunt de dades oficials de la UEFA i selecciona un atribut del conjunt de dades. Calcula el p-valor i digues si rebutja la hipòtesi nul-la agafant un alfa de 5%

```python
#H0: la mostra té una distribució gaussiana
#H1: la mostra no té una distribució gaussiana
#trio latribut de partits guanyats per veure si té una distribució gaussiana

from scipy.stats import shapiro
data = uefa[['W']]
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' %(stat, p))
if p > 0.05:
      print('Probablement tenen una distribució gaussiana, acceptem hipotesi nul-la')
else:
      print('Probablement no tenen una distribució gaussiana, rebutjem hipotesi nul-la')
```

## Exercici 2
Amb el mateix dataset selecciona dos altres atributs del conjunt de dades. Calcula els p-valors i digues si rebutgen la hipòtesi nul-la agafant un alfa de 5%.

```python
# H0: les mitjes de les mostres són iguals
# H1: les mitjes de les mostres són desiguals
# trio els atributs de gols ficats i gols en contra.

from scipy.stats import ttest_ind
data1 = uefa[['GF']]
data2 = uefa[['GA']]
stat, p = ttest_ind(data1, data2)
print('stat=%.3f, p=%.3f' % (stat,p))
if p > 0.05:
    print('les mitjes de les mostres són iguals, acceptem hipotesi nul-la')
else:
    print('les mitjes dee les mostres són desiguals, rebutjem hipotesi nul-la')
```

## Exercici 3

Continua amb el conjunt de dades adjunt i selecciona tres atributs del conjunt de dades. Calcula el p-valor i digues si rebutja la hipòtesi nul-la agafant un alfat del 5%

```python
# H0: les mitjes de les mostres són iguals
# H1: les mitjes de les mostres són desiguals
# per fer aquest exercici triu l'anàlisi ANOVA perque són més de dos atributs
# trio els atributs de gols marcats, partits guanyats i 
from scipy.stats import f_oneway
data1 = uefa[['D']]
data2 = uefa[['L']]
data3 = uefa[['xGA']]
stat, p = f_oneway(data1, data2, data3)
print('stat=%.3f, p=%.3f' % (stat, p))

if p > 0.5:
    print('Probablement tenen la mateixa distribució, acceptem hipotesi nul-la')
else:
    print('Probablement tenen una distribució diferent, rebutjem hipotesi nul-la')
```
